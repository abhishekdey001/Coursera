### The following code calculates speed w.r.t distance and time 

```python
def cal_speed(distance, time):
    speed = distance / time
    return speed

distance_in_meter = 100
time_in_sec = 20

speed_in_meter_per_sec = cal_speed(distance_in_meter, time_in_sec)

print("Speed (m/sec) = ", speed_in_meter_per_sec)
```
